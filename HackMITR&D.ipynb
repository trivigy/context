```python
def plot_learning_curve(model, err_func=explained_variance_score, N=15, n_runs=10, n_sizes=50, ylim=None):
    sizes = np.linspace(5, N, n_sizes).astype(int)
    train_err = np.zeros((n_runs, n_sizes))
    validation_err = np.zeros((n_runs, n_sizes))
    for i in range(n_runs):
        for j, size in enumerate(sizes):
            xtrain, xtest, ytrain, ytest = train_test_split(
                X_test, y_test, train_size=size, random_state=i)
            #Train on only the first `size` points
            model.fit(X_train.toarray(), y_train)
            validation_err[i, j] = err_func(y_test, model.predict(X_test.toarray()))
            train_err[i, j] = err_func(y_train, model.predict(X_train.toarray()))

    plt.plot(sizes, validation_err.mean(axis=0), lw=2, label='validation')
    plt.plot(sizes, train_err.mean(axis=0), lw=2, label='training')

    plt.xlabel('traning set size')
    plt.ylabel(err_func.__name__.replace('_', ' '))
    
    plt.grid(True)
    
    plt.legend(loc=0)
    
    plt.xlim(0, N-1)
    
    if ylim:
        plt.ylim(ylim)
```

```python
plot_learning_curve(GaussianNB(),err_func=mean_squared_error)
```
